#### 1. Import libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

#### 2. Read images and check shape

```python
img1 = cv2.imread("logo.png")
img2 = cv2.imread("messi.png")
print(img1.shape)
print(img2.shape)
```

#### 3. Resize the img1 for matching the img2 shape

```python
img1 = cv2.resize(img1, (img2.shape[1], img2.shape[0]))
print(img1.shape)
```

#### Theory of image blending

## dst = α . img1 + β . img2 + γ

#### 4. Image Blending

```python
# here the value of α = 0.3 and β = 0.7 and γ = 0
dst = cv2.addWeighted(img1, 0.7, img2, 0.3, 0)

# Color is converted from BGR to RGB to visualize in matplotlib 
dst = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(dst)
plt.show()
```
